### **Métodos em Python**

Métodos são funções associadas a objetos que permitem manipular ou obter informações sobre eles.

Exemplos de métodos em strings:

````python
texto = "Olá, mundo! tudo bem mundo"
print(texto)
print("upper:", texto.upper())  # Converte para maiúsculas
print("lower:", texto.lower())  # Converte para minúsculas
print("replace:", texto.replace("mundo", "Python"))  # Substitui palavras
````

````python
"ola mundo".upper() + " tudo bem " + str(100)
````

### **Definição de Listas e Mapas**

Listas armazenam múltiplos valores em uma sequência ordenada.

````python
frutas = ["maçã", "banana", "laranja"]
print(frutas)
print(frutas[0])  # Acessa o primeiro elemento
print(len(frutas))
````

````python
# acessa ao ultimo elemento da lista frutas
##### YOUR CODE HERE #######
frutas[-1]
############################
````

Dicionários (Mapas) armazenam pares chave-valor.

````python
aluno = {"nome": "Ana", "idade": 25,
         "curso": "Python", "lista": [1,4,8]}
print(aluno["nome"])  # Acessa o valor associado à chave "nome"
print(aluno["idade"])  # Acessa o valor associado à chave "idade"
print(aluno["lista"])
#print(aluno);
````

Iteração sobre listas e dicionários:

````python
for i in range(0,len(frutas)):
   print(i, frutas[i])
print("-------------")
for iten_fruta in frutas:
    print(iten_fruta)
````

````python
print(aluno.items())

for chave, valor in aluno.items():
    #print(chave, valor)
    print(f"{chave}: {valor}")
````

### **Função em Python**

````python
def soma(a, b):
    """
    Esta função recebe dois números e retorna a soma deles.

    Parâmetros:
    a (int ou float): O primeiro número.
    b (int ou float): O segundo número.

    Retorna:
    int ou float: A soma de a e b.
    """
    valor_soma = a + b
    return valor_soma

# Exemplo de uso da função
resultado = soma(5, 10)
print(f"A soma de 5 e 10 é: {resultado}")
````
