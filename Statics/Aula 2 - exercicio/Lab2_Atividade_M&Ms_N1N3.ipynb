#### **3- Frequência acumulada ($F_i$)**
Mostra a quantidade acumulada de M&Ms até uma determinada cor, ordenadas por alguma classificação (utilizar: verde, vermelho, azul, amarelo, roxo, laranja). A frequência acumulada está definida por $F_i$,

$$
F_i = f_1 + f_2 + f_3 + \dots + f_n \ \ \ \text{,  ou seja} \\
F_{azul} = f_{verde} + f_{vermelho} + f_{azul}
$$

e.g.,



Crie uma função que receba a lista de frequências simples (contagens) e retorne uma lista de frequência acumulada ($F_i$).


```python
def freq_acumulada(lista_freq_simples):
  n = sum(lista_freq_simples) # soma do número total de M&Ms
  lista_freq_acumulada = [0]*len(lista_freq_simples) # iniciar uma lista preenchida com zeros
  ######## YOUR CODE HERE ########
  for i in range(len(lista_freq_simples)):
    if i == 0:
      lista_freq_acumulada[i] = lista_freq_simples[i]
    else:
      lista_freq_acumulada[i] = lista_freq_simples[i] + lista_freq_acumulada[i-1]


  ################################
  return lista_freq_acumulada
fi = [13, 10, 6, 5, 5, 12]

fac = freq_acumulada(fi) # chamar a função e enviar o parametro fi
print(fac)
```

Verificando se a lista de frequência acumulada esta correta



```python
# Cálculo da frequência acumulada em Pandas
df['frequencia_relativa'] = df['frequencia_simples'] / df['frequencia_simples'].sum()

df['frequencia_acumulada'] = df['frequencia_simples'].cumsum()

verificar_valores(fac, df['frequencia_acumulada'].tolist())
```

#### **5- Média aritmética**
É o valor obtido somando todos os elementos de um conjunto de dados e dividindo essa soma pelo número total de elementos. Está definida por $\bar{X}$ (para amostra) e $\mu$ (para população).

$$
\bar{X} = \sum\frac{x_i}{n}
$$

Criar uma função que que receba a lista de frequências simples (contagens) e calcule a média aritmética dos M&Ms.

```python
def media_aritmetica(lista_freq_simples):
  n = len(lista_freq_simples) # numero de cores de M&Ms
  ######## YOUR CODE HERE ########
  soma = 0
  for i in range(n):
    soma += lista_freq_simples[i] / n
  ################################
  return soma

media = media_aritmetica(fi) # chamar a função e enviar o parametro fi
print(media)
```

Verificando se a média aritmética esta correta

```python
# Cálculo da media aritmetica em pandas
df_media = df['frequencia_simples'].mean()

verificar_valores(media, df_media)
```

```python
def moda(lista_freq_simples, rotulo_cores):
  moda = []
  ######## YOUR CODE HERE ########
  max_freq = max(lista_freq_simples)
  for i in range(len(lista_freq_simples)):
    if lista_freq_simples[i] == max_freq:
      moda.append(rotulo_cores[i])
  ################################
  return moda

label_cores = ['Verde', 'Vermelho', 'Azul', 'Amarelo', 'Roxo', 'Laranja']
Mo = moda(fi, label_cores) # chamar a função e enviar o parametro fi e lista de cores
print(Mo)
```
