```python
# Passo 1 - Criando dados para manusearmos
#T1a
import pandas as pd

# Criando o DataFrame
dados = {
    'Nivel_Atividade_Fisica': list(range(20)),
    'Genero': ['Homem', 'Mulher'] * 10,
    'Nivel_Risco': ['Alto'] * 6 + ['Médio'] * 8 + ['Baixo'] * 6,
    'Risco_Doenca_Cardiaca': [
        96.9, 88.5, 87.3, 87.4, 74.1, 67.5, 71.3, 62.0, 50.8, 50.7,
        40.6, 35.2, 33.6, 17.5, 13.3, 14.9, 7.4, 9.8, 0.5, 0.1
    ]
}

df = pd.DataFrame(dados)
df
```

```python
# Exercio 01
#T1a
import matplotlib.pyplot as plt

# Filtrar pessoas com risco >= 60
doenca = df[df['Risco_Doenca_Cardiaca'] >= 60]

# Contar quantos de cada gênero
contagem_genero = doenca['Genero'].value_counts()

# Plotar
contagem_genero.plot(kind='bar', color=['blue', 'pink'])
plt.title('Frequência de Gênero com Doença Cardíaca (≥60%)')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.show()
```

```python
# Exercicio 02
#T2a

# Frequência relativa
proporcao_nivel_risco = df['Nivel_Risco'].value_counts(normalize=True)

# Gráfico de pizza
proporcao_nivel_risco.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proporção por Nível de Risco')
plt.ylabel('')
plt.show()
```

```python
# Exercicio 03

total = len(df)
# Filtrar vendedores que venderam mais de 15 unidades
vendedores_mais_15 = df[df["Unidades Vendidas"] > 15].shape[0]
# Calcular a probabilidade
probabilidade = vendedores_mais_15 / total
# Exibir resultado
print(f"Vendedores com mais de 15 unidades vendidas: {vendedores_mais_15}")
print(f"Total de vendedores: {total}")
print(f"Probabilidade: {probabilidade:.2f}")

```

```python
# Exercicio 05
#T5a

import numpy as np
import matplotlib.pyplot as plt

x = df['Nivel_Atividade_Fisica']
y = df['Risco_Doenca_Cardiaca']

# Cálculos de mínimos quadrados
x_media = np.mean(x)
y_media = np.mean(y)

# β1 = soma((xi - x̄)(yi - ȳ)) / soma((xi - x̄)^2)
beta1 = np.sum((x - x_media) * (y - y_media)) / np.sum((x - x_media)**2)
# β0 = ȳ - β1 * x̄
beta0 = y_media - beta1 * x_media

print(f'β0 (intercepto): {beta0:.2f}')
print(f'β1 (inclinação): {beta1:.2f}')

#calculo x̄ e ȳ
#(x̄) = (0 + 1 + 2 + ... + 19) / 20 = 190 / 20 = 9.5
#Soma dos y = 919.5
#média de y (ȳ) = 919.5 / 20 = 45.975

#β1 = -1626.85 / 1330 ≈ -1.224

#β0 = ȳ - β1 * x̄
#β0 = 45.975 - (-1.224 * 9.5)
#β0 = 45.975 + 11.628 = 57.603

#EQUAÇÃO FINAL DA REGRESSÃO:
#risco_doenca = 57.60 - 1.22 * nivel_atividade

# Gráfico com reta
plt.scatter(x, y, label='Dados')
plt.plot(x, beta0 + beta1 * x, color='red', label='Regressão Linear')
plt.xlabel('Nível de Atividade Física (horas)')
plt.ylabel('Risco de Doença Cardíaca (%)')
plt.title('Regressão Linear Simples')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Exercicio 06
#T6a

from sklearn.linear_model import LinearRegression

# Transformar para arrays 2D
X = x.values.reshape(-1, 1)
Y = y.values

modelo = LinearRegression()
modelo.fit(X, Y)

print(f'β0 (intercepto): {modelo.intercept_:.2f}')
print(f'β1 (inclinação): {modelo.coef_[0]:.2f}')

# Gráfico
plt.scatter(x, y, label='Dados')
plt.plot(x, modelo.predict(X), color='purple', label='ML Regressão Linear')
plt.xlabel('Nível de Atividade Física (horas)')
plt.ylabel('Risco de Doença Cardíaca (%)')
plt.title('Regressão Linear (ML)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercicio 07
#T7a

# Valores a prever
novos_valores = np.array([[7.5], [15.3]])
previsoes = modelo.predict(novos_valores)

# Mostrar previsões
for i, horas in enumerate([7.5, 15.3]):
    print(f'Atividade Física: {horas}h => Risco: {previsoes[i]:.2f}%')

# Gráfico com os pontos previstos
plt.scatter(x, y, label='Dados')
plt.plot(x, modelo.predict(X), color='purple', label='Regressão')
plt.scatter([7.5, 15.3], previsoes, color='red', label='Previsões', marker='x', s=100)
plt.xlabel('Nível de Atividade Física (horas)')
plt.ylabel('Risco de Doença Cardíaca (%)')
plt.title('Previsão com Modelo ML')
plt.legend()
plt.grid(True)
plt.show()
```
