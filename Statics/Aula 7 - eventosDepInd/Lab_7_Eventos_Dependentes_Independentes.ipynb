#### **1- O que é Probabilidade?**

A probabilidade é o ramo da matemática que estuda a chance de ocorrência de eventos em experimentos aleatórios. Ela é usada para quantificar a incerteza e prever a possibilidade de diferentes resultados.

$$
P(A) = \frac{\text{número de casos favoráveis a } A}{\text{número total de casos possíveis}}
$$
<br>
$$
P(A) = \frac{n(A)}{n(S)}
$$

A onde:
* $P(A)$: probabilidade do evento $A$
* $n(A)$: número de casos favoráveis (evento $A$)
* $n(S)$: número total de casos possíveis (espaço amostral)

#### **2- Espaço Amostral e Evento**

#### **Espaço Amostral**

O espaço amostral (denotado por $S$ ou $\Omega$) é o conjunto de todos os possíveis resultados de um experimento aleatório.


**Exemplo 1:** <br>
Ao lançar um dado de 6 faces, o espaço amostral é:

$$S = \{1, 2, 3, 4, 5, 6\}$$


**Exemplo 2:** <br>
 Ao lançar uma moeda, o espaço amostral é:

$$S = \{\text{cara}, \text{coroa} \}$$


#### **Evento**

Um evento é qualquer subconjunto do espaço amostral. Ou seja, um evento é um ou mais resultados possíveis do experimento.


Evento Simples: Contém apenas um resultado.

**Exemplo 1:**<br>
 "Sair o número 4 ao lançar um dado"
Evento: $\{4\}$<br>
Espaço amostral:$\{1,2,3,4,5,6\}$

**Exemplo 2:**<br>
"Seleção de uma pessoa em uma lista de chamada"

Evento: Ser sorteado o aluno João.<br>
Espaço amostral: Todos os alunos da lista.

### **3- Eventos Independentes**
Dois eventos são independentes se a ocorrência de um não afeta a ocorrência do outro.

Equação para eventos independentes
Se $A$ e $B$ são independentes:

<br>
$$
P(A \cap B) = P(A) \times P(B)
$$

<br>



Por exemplo: Lançar uma moeda e lançar um dado: o resultado de um não influencia o outro.

### **4- Eventos Dependentes**
Dois eventos são dependentes se a ocorrência de um afeta a ocorrência do outro.

Equação para eventos dependentes
Se $A$ e $B$ são dependentes:

<br>
$$
P(A \cap B) = P(A) \times P(B|A)
$$
<br>

onde $P(B|A)$ é a probabilidade de $B$ ocorrer dado que $A$ já ocorreu.



Por exemplop: A primeira seja azul e a segunda seja verde.

#### **Interseção de eventos independentes, $P(A) \times P(B)$**

**T1-** Qual a probabilidade de uma peça ser produzida no turno da manhã e ser aprovada no controle de qualidade?

```python
#### YOUR CODE HERE ####
contador_turno_manha = 0
for i in range(len(df)):
    if df["Turno"][i] == "Manhã":
        contador_turno_manha += 1

contador_peca_aprovada = 0
for i in range(len(df)):
    if df["Peça Aprovada?"][i] == "Sim":
        contador_peca_aprovada += 1


prob_interseccao_manha_aprovada = (contador_turno_manha / len(df)) * (contador_peca_aprovada / len(df))

print(prob_interseccao_manha_aprovada)
```

**T2-** Qual a probabilidade de uma peça ser produzida no turno da tarde e ser operada por um operador treinado?

```python
#### YOUR CODE HERE ####
contador_turno_tarde = 0
for i in range(len(df)):
    if df["Turno"][i] == "Tarde":
        contador_turno_tarde += 1

contador_operador_treinado = 0
for i in range(len(df)):
    if df["Operador Treinado?"][i] == "Sim":
        contador_operador_treinado += 1


prob_interseccao_tarde_operador_treinado = (contador_turno_tarde / len(df)) * (contador_operador_treinado / len(df))

print(prob_interseccao_tarde_operador_treinado)
```

**T3-** Qual a probabilidade de uma peça ser produzida no turno da noite e ser feita na máquina M1?

```python
#### YOUR CODE HERE ####
contador_turno_noite = 0
for i in range(len(df)):
    if df["Turno"][i] == "Noite":
        contador_turno_noite += 1

contador_maquina = 0
for i in range(len(df)):
    if df["Máquina Usada"][i] == "M1":
        contador_maquina += 1


prob_produzida_noite_maquina = (contador_turno_noite / len(df)) * (contador_maquina / len(df))

print(prob_produzida_noite_maquina)
```

**T4-** Qual a probabilidade de uma peça ser aprovada no controle de qualidade e ser feita na máquina M2?

**T5-** Qual a probabilidade de uma peça ser feita por um operador treinado e não ter sido reprovada anteriormente?

#### **Interseção de eventos dependentes, $P(A) \times P(B|A)$**

**T6-** Qual a probabilidade de uma peça ser produzida no turno da manhã e, dado que foi produzida nesse turno, ser reprovada no controle de qualidade?

```python
#### YOUR CODE HERE ####

contador_turno_manha = 0
for i in range(len(df)):
    if df["Turno"][i] == "Manhã":
        contador_turno_manha += 1

contador_reprovada_manha = 0
for i in range(len(df)):
    if df["Turno"][i] == "Manhã" and df["Peça Reprovada Anteriormente?"][i] == "Sim":
        contador_reprovada_manha += 1

# Probabilidade de ser produzida no turno da manhã e reprovada
prob_interseccao_manha_reprovada = (contador_turno_manha / len(df)) * (contador_reprovada_manha / contador_turno_manha)

print(prob_interseccao_manha_reprovada)
```

**T7-** Qual a probabilidade de uma peça ser feita na máquina M3 e, dado que foi feita nessa máquina, ser operada por um operador não treinado?

```python
#### YOUR CODE HERE ####
contador_maquinaM3 = 0
for i in range(len(df)):
    if df["Máquina Usada"][i] == "M3":
        contador_maquinaM3 += 1

contador_nao_treinadoM3 = 0
for i in range(len(df)):
    if df["Máquina Usada"][i] == "M3" and df["Operador Treinado?"][i] == "Não":
        contador_nao_treinadoM3 += 1

# Probabilidade de ser feita na máquina M3 e operada por um operador não treinado
prob_interseccao_m3_nao_treinado = (contador_maquinaM3 / len(df)) * (contador_nao_treinadoM3 / contador_maquinaM3)

print(prob_interseccao_m3_nao_treinado)
```

**T8-** Qual a probabilidade de uma peça ser aprovada no controle de qualidade e, dado que foi aprovada, ter sido feita na máquina M1?

```python
#### YOUR CODE HERE ####

contador_peca_aprovada = 0
for i in range(len(df)):
    if df["Peça Aprovada?"][i] == "Sim":
        contador_peca_aprovada += 1

contador_aprovadaM1 = 0
for i in range(len(df)):
    if df["Peça Aprovada?"][i] == "Sim" and df["Máquina Usada"][i] == "M1":
        contador_aprovadaM1 += 1

# Probabilidade de ser aprovada e feita na máquina M1
prob_condicional_aprovadaM1 = (contador_aprovadaM1 / len(df)) / (contador_peca_aprovada / len(df))

print(prob_condicional_aprovadaM1)

```

**T9-** Qual a probabilidade de uma peça ser feita por um operador não treinado e, dado isso, ter sido reprovada anteriormente?

```python
#### YOUR CODE HERE ####

contador_nao_treinado = 0
for i in range(len(df)):
    if df["Operador Treinado?"][i] == "Não":
        contador_nao_treinado += 1

contador_nao_treinado_reprovada = 0
for i in range(len(df)):
    if df["Operador Treinado?"][i] == "Não" and df["Peça Reprovada Anteriormente?"][i] == "Sim":
        contador_nao_treinado_reprovada += 1

# Probabilidade de ser feita por um operador não treinado e reprovada
prob_condicional_nao_treinado_reprovada = (contador_nao_treinado_reprovada / len(df)) / (contador_nao_treinado / len(df))

print(prob_condicional_nao_treinado_reprovada)
```
